```python
import pandas as pd
from preprocessing import DataFramePreprocessor
from sklearn.datasets import load_wine
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, OneHotEncoder
```

```python
titanic = pd.read_csv('../data/titanic.csv')
titanic = titanic.iloc[:20, :]
titanic
```

```python
lbl = LabelEncoder()
```

```python
dfp = DataFramePreprocessor()
```

```python
ohe = OneHotEncoder()
```

```python
mms = MinMaxScaler()
```

```python
numeric_cols = ["Parch", "Age", "Fare"]
onehot_col = ["Sex"]
label_col = ["Pclass"]
dfp = DataFramePreprocessor()
df, tfs = dfp.fit_transform_single_transformer(titanic, lbl, onehot_col)
df
```

```python
df, tfs = dfp.fit_transform_multiple_transformer(titanic, [mms, ohe, lbl], [numeric_cols, onehot_col, label_col])
df
```

```python
df, tfs = dfp.fit_transform_multiple_transformer(titanic[0:10], [mms, ohe, lbl], [numeric_cols, onehot_col, label_col])
df
```

```python
col = [[k] for k in tfs]
col
```
